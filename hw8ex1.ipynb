```python
import pandas as pd

file_data = pd.read_csv('amazon_train_df.csv')
```

```python
#change the name of columns
file_data.columns = ['review_title', 'rating', 'review_date', 'customer_name', 'review']
```

```python
#Cleaning
#Removing HTML tag
from bs4 import BeautifulSoup

def remove_html(text_data):
    if isinstance(text_data, str):
        soup = BeautifulSoup(text_data, 'lxml')
        return soup.get_text()
    else:
        return text_data

# processed_text = remove_html(str_data)
# print(processed_text)
```

```python
for column in file_data.columns:
    file_data[column] = file_data[column].apply(remove_html)
```

```python
#Removing punctuation
import string
print('Punctuation: ', string.punctuation)

def remove_punctuation(text):
    if not isinstance(text, str):
        return text

    sent = []
    for t in text.split():
        no_punct = "".join([c for c in t if c not in string.punctuation])
        sent.append(no_punct)

    sentence = " ".join(s for s in sent)
    return sentence

# rmv_punc_sentence = remove_punctuation(processed_text)
#print(rmv_punc_sentence)
```

```python
#lower case
def to_lower_case(text_data):
    if isinstance(text_data, str):
        text_data = text_data.lower()
        return text_data
    else:
        return text_data
```

```python
for column in file_data.columns:
  file_data[column] = file_data[column].apply(remove_punctuation)
  file_data[column] = file_data[column].apply(to_lower_case)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def tokenize_and_lemmatize(text):
    if not isinstance(text, str):
        return text

    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc]
    return " ".join(lemmas)

for column in file_data.columns:
    file_data[column] = file_data[column].apply(tokenize_and_lemmatize)
```

```python
import spacy
import nltk
from nltk.corpus import stopwords

nlp = spacy.load('en_core_web_sm')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    if not isinstance(text, str):
        return text

    doc = nlp(text)
    tokens = [token.text for token in doc if token.text not in stop_words]
    return " ".join(tokens)
```

```python
import pandas as pd
from collections import Counter

review_column = file_data['review']

top_five_letters_per_review = []

for review in review_column:
    review_letters = ''.join(review).lower()
    review_letters = ''.join(letter.upper() for letter in review_letters if letter.isalpha())
    letter_counts = Counter(review_letters)
    top_five_letters = [letter for letter, count in letter_counts.most_common(5)]
    top_five_letters_per_review.append(top_five_letters)

for i, top_five_letters in enumerate(top_five_letters_per_review, 1):
    print(f"Review {i}: {', '.join(top_five_letters)}")
```
